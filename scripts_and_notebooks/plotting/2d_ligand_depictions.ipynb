```python
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
highlightColor = (0.5, 0.5, 0.5)
size = (300, 300)
```

```python
thiane = Chem.AddHs(Chem.MolFromSmiles("C1CCSCC1"))
img = Draw.MolToImage(
    thiane, size=size, highlightAtoms=[3], highlightColor=highlightColor
)
img.save("plots/thiane_alt.png")
img
```

```python
phchoh = Chem.AddHs(Chem.MolFromSmiles("[PH]=[CH]-[OH]"))
img = Draw.MolToImage(
    phchoh, size=size, highlightAtoms=[0], highlightColor=highlightColor
)
img.save("plots/[PH]=[CH]-[OH]_alt.png")
img
```

```python
bipyrimidine = Chem.AddHs(Chem.MolFromSmiles("C1=CN=C(N=C1)C2=NC=CC=N2"))
img = Draw.MolToImage(
    bipyrimidine, size=size, highlightAtoms=[4, 7], highlightColor=highlightColor
)
img.save("plots/bipyrimidine_alt.png")
img
```
