```python
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
options = Draw.MolDrawOptions()
options.atomHighlightsAreCircles = False
options.setBackgroundColour((1.0, 1.0, 1.0, 0.0))
options.setHighlightColour((0.5, 0.5, 0.5))
options.highlightRadius = 0.4  # default 0.3
options.scaleBondWidth = True
options.bondLineWidth = 4.0
```

```python
size = (400, 400)
add_Hs = False
```

```python
thiane = Chem.MolFromSmiles("C1CCSCC1")
if add_Hs:
    thiane = Chem.AddHs(thiane)
img = Draw.MolToImage(
    thiane,
    size=size,
    highlightAtoms=[3],
    options=options,
)
img.save("plots/thiane_alt.png")
img
```

```python
phchoh = Chem.MolFromSmiles("[PH]=[CH]-[OH]")
if add_Hs:
    phchoh = Chem.AddHs(phchoh)
options.rotate = 180
img = Draw.MolToImage(phchoh, size=size, highlightAtoms=[0], options=options)
options.rotate = 0
img.save("plots/[PH]=[CH]-[OH]_alt.png")
img
```

```python
bipyrimidine = Chem.MolFromSmiles("C1=CN=C(N=C1)C2=NC=CC=N2")
if add_Hs:
    bipyrimidine = Chem.AddHs(bipyrimidine)
img = Draw.MolToImage(bipyrimidine, size=size, highlightAtoms=[4, 7], options=options)
img.save("plots/bipyrimidine_alt.png")
img
```
